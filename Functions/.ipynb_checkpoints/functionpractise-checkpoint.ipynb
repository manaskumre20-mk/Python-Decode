```python
# wap function that return whether a given year is leap or not ?
def leap_year(year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
```

```python
year1=int(input("enter any number: "))
if leap_year(year1):
    print(year1,"leap year")
else:
    print(year1,"not a leap year")
```

```python
# wap fuction that return fibonacci series upto nth term?
def fibo(n):
    if n==1:
        return [0]
    elif n==2:
        return [0,1]
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib
```

```python
fibo(7)
```

```python
# wap function that return n term of fibonacci series?
def fibo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib[-1]
```

```python
fibo(7)
```

```python
# wap function that return prime number between 1 and 100?
def prime(start,stop):
    l=[]
    for i in range(start,stop+1):
        if i>1:
            for j in range(2,i):
                if i%j==0:
                    break
            else:
                l.append(i)
    return l
```

```python
prime(1,20)
```

```python
#wap function that return largest among three number?
def max_of_two(x,y):
    if x>y:
        return x
    return y

def max_of_three(a,b,c):
    return max_of_two(a,max_of_two(b,c))
```

```python
max_of_three(10,100,20)
```

```python
def arm(n):
    a=len(str(n))
    old=n
    sm=0
    while n!=0:
        dig=n%10
        sm=sm+(dig**a)
        n=n//10
    return sm==old
```

```python
n1=int(input("enter any number: "))
if arm(n1):
    print(n1,"armstrong number")
else:
    print(n1,"not an armstrong number")
```

```python
def armstrong(start,stop):
    l=[]
    for i in range(start,stop+1):
        a=len(str(i))
        old=i
        sm=0
        while i!=0:
            dig=i%10
            sm=sm+(dig**a)
            i=i//10
        if old==sm:
            l.append(old)
    return l
```

```python
armstrong(1,2000)
```
