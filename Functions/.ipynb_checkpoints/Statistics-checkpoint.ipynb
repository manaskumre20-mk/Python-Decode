```python
age=np.array([3,23,24,23,25,26,27,27,24,27,21,23,28,29,26,25,24,22,21,26,25,56])
print(age)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(x=age,color="purple")
plt.show()
```

```python
print(np.mean(age))
print(np.median(age))
```

```python
Q1=np.percentile(age,25)
print(Q1)
```

```python
Q2=np.quantile(age,0.50)
print(Q2)
```

```python
Q3=np.quantile(age,0.75)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*IQR
print(lower_fence)
```

```python
upper_fence=Q3+1.5*IQR
print(upper_fence)
```

```python
l=[]
for i in age:
    if i>upper_fence or i<lower_fence:
        l.append(i)
print(l)
```
