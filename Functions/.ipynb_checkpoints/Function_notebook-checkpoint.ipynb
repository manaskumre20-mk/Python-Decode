```python
#without parameter without returntype 
def sum():
    a=int(input("enter any number: "))
    b=int(input("enter any number: "))
    c=a+b
    print(c)
```

```python
#without parameter with returntype
def diff():
    a=int(input("enter any number: "))
    b=int(input("enter any number: "))
    c=a-b
    return c
    
```

```python
print(diff())
```

```python
#with parameter without returntype
def mult(a,b):
    c=a*b
    print(c)
```

```python
mult(10,2)
```

```python
#with parameter with returntype
def div(a,b):
    c=a/b
    return c
```

```python
n1=int(input("enter any number: "))
n2=int(input("enter any number: "))
div(n1,n2)
```

```python
#optional parameter
def sum(a,b=10):
    c=a+b
    return c
```

```python
sum(50)
```

```python
def sort_list(l,reverse=False):
    if reverse==False:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]<l[j]:
                    l[i],l[j]=l[j],l[i]
                    return l
    else:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]>l[j]:
                    l[i],l[j]=l[j],l[i]
                    return l
        
```

```python
def leap_year(year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
```

```python
year1=int(input("enter any year: "))
if leap_year(year1):
    print(year1,"leap year")
else:
    print(year1,"not a leap year")
```
