```python
import numpy as np
arr=np.array([1,2,3,4])
print(arr)
arr1=arr+3
print(arr1)
```

```python
arr1=np.array([1,2,3,4])
print(arr1)
arr2=np.array([[10],[20],[30],[40]])
print(arr2)
arr=arr1+arr2
print(arr)
```
