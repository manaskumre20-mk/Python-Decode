```python
#wap function that returns sroted list eihter in asc order or in desc order on the basis of optional parameter i.e. (reverse=False)?
#input - l=[19,1,23,17,16,3]
#output- reverse=False - [1,3,16,17,19,23]
def sorted(l,reverse=False):
    if reverse==False:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]>l[j]:
                    l[i],l[j]=l[j],l[i]
        return l
    else:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]<l[j]:
                    l[i],l[j]=l[j],l[i]
        return l
```

```python
l=[19,1,23,17,16,3]
sorted(l,reverse=True)
```

```python
#recursion function
#wap recursion function that return sum of first n natural number?
def sum(n):
    if n==1:
        return n
    else:
        return n+sum(n-1)
```

```python
sum(10)
```

```python
#wap recursion function that return factorial of nth number?
def factorial(n):
    if n==1:
        return n
    else:
        return n*factorial(n-1)
```

```python
factorial(5)
```
