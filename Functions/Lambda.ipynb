```python
# lambda arguments:expression
# A function without name is lambda function
# it is a small anonymous function 
# A lambda function can take any number of arguments but can only have one expression.
# it is used with high order function like
#apply,map,reduce or filter.
```

```python
a=lambda x:x**3
print(a(2))
```

```python
a=lambda x,y:x if x>y else y
print(a(10,100))
```

```python
a=lambda x,y,z:x if x>y else y if y>z else z
print(a(10,100,200))
```

```python
#map
#map(function,iterables)
l1=[1,2,3,4,5]
l=[]
for i in l1:
    l.append(i**3)
print(l)
print([i**3 for i in l1])
print(list(map(lambda i:i**3,l1)))
```

```python
l1=[5,1,2,3,4,6]
l2=[2,6,5,4,3,1]
#[25,1,36,81,64,6]
l=[]
for i,j in zip(l1,l2):
    l.append(i**j)
print(l)
print([i**j for i,j in zip(l1,l2)])
print(list(map(lambda i,j:i**j,l1,l2)))
```
