### Join Two Array

```python
import numpy as np
arr1=np.array([1,2,3,4])
print(arr1)
arr2=np.array([5,6,7,8])
print(arr2)
print()
arr=np.concatenate((arr1,arr2))
print(arr)
```

```python
arr1=np.array([[1,2,3,4,5]])
print(arr1)
arr2=np.array([[6,7,8,9,10]])
print(arr2)
print()
arr_new=np.concatenate((arr1,arr2))
print(arr_new)
arr_new1=np.concatenate((arr1,arr2),axis=1)
print(arr_new1)
```

```python
arr1=np.array([[1,2,3,4,5]])
print(arr1)
arr2=np.array([[6,7,8,9,10]])
print(arr2)
print()
arr_new=np.stack((arr1,arr2))
print(arr_new)
print()
arr_new_v=np.vstack((arr1,arr2))
print(arr_new_v)
print()
arr_new_h=np.hstack((arr1,arr2))
print(arr_new_h)
print()
arr_new_d=np.dstack((arr1,arr2))
print(arr_new_d)
```

```python
#split() - equal division
arr=np.array([10,20,30,40,50,60])
arr1=np.split(arr,3)
print(arr1)
```

```python
#array_split() - unequal division 
arr=np.array([10,20,30,40,50,60,70,80])
print(arr)
arr1=np.array_split(arr,3)
print(arr1)
```
