```python
#Set
#collection of element
#denoted by {} with csv
#heterogenous in nature, unordered, unindexed and mutable
#stores only unique element.
s=set() #set constructor
print(s)
print(type(s))
```

```python
s=set([10,20,30,"hello",10.5,10,20,10])
print(s)
```

```python
#set(iterable object)
s=set((10,20,30,"hello",10.5,10,20,10))
print(s)
```

```python
s={10,20,30,"hello",10.5,10,20,10}
print(s)
print(type(s))
#print(s[0]) #unindexed
s.add(100) #mutable
print(s)
```

```python
#direct iteration
s={10,20,30,"hello",10.5,10,20,10}
for i in s:
    print(i)
```

```python
s1={1,2,3,5,7,9}
s2={11,12,16,7,3}
#union (|)
print(s1.union(s2))
print(s1|s2)
#intersection (&)
print(s1.intersection(s2))
print(s1&s2)
#difference (-)
print(s1.difference(s2))
print(s1-s2)
#symmetric_difference (^)
print(s1.symmetric_difference(s2))
print(s1^s2)
```

```python
#pop()
#setobj.pop()
s={10,20,30,"hello",10.5}
print(s.pop())
print(s)
```

```python
#setobj.remove(value)
s={10,20,30,"hello",10.5}
s.remove("bhopal")
print(s)
```

```python
#setobj.discard(value)
s={10,20,30,"hello",10.5}
s.discard("bhopal")
print(s)
```

```python
#clear()
#setobj.clear()
s={10,20,30,"hello",10.5}
s.clear()
print(s)
```
