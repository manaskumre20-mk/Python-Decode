```python
#dictionary
#dictobj.get(key)
d={"name":["xyz","abc","def"],"subject":["da","ds","de"],"fees":[30000,40000,50000]}
print(d.get("name"))
for i in d.get("name"):
    print(i)
```

```python
#keys()
#dictobj.keys()
d={"name":["xyz","abc","def"],"subject":["da","ds","de"],"fees":[30000,40000,50000]}
print(d.keys())
for i in d.keys():
    print(i)
```

```python
#values()
#dictobj.values()
d={"name":["xyz","abc","def"],"subject":["da","ds","de"],"fees":[30000,40000,50000]}
print(d.values())
for i in d.values():
    print(i)
```

```python
#items()
#dictobj.items()
d={"name":["xyz","abc","def"],"subject":["da","ds","de"],"fees":[30000,40000,50000]}
for i,j in d.items():
    print(i,j)
```

```python
l=[1,4,3,2,5,1,5,6,7,7,7,3,2,9,8,7,7]
d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print(d)
max=0
for i in d:
    for j in d:
        if d[i]>d[j]:
            max=d[i]
for i,j in d.items():
    if j==max:
        print(i)
```

```python
l=[1,4,3,2,5,1,5,7,7,7,3,2,9,8,7,7]
sum=0
count=0
mean=0
for i in l:
    sum=sum+i
    count=count+1
    mean=sum/count
print(mean)
    
```

```python
l=[1,4,3,2,5,1,5,6,7,7,7,3,2,9,8,7,7]
l1=sorted(l)
print(l1)
if len(l)%2==0:
    median=(l1[len(l1)//2-1]+l1[len(l1)//2])/2
else:
    median=l1[len(l)//2]
print(median)
```

```python
words=["data","science","AI","python","data","code","python"]
d={}
for i in words:
    if len(i)>3:
         if i in d:
            d[i]+=1
         else:
            d[i]=1
print(d)
```

```python
marks={
    "Ankit":90,
    "Ravi":85,
    "Neha":90,
    "Aman":80}
d={}
for i,j in marks.items():
    if j not in d:
        d[j]=[]
    d[j].append(i)
print(d)
```

```python
nums=[10,15,21,22,30,33,40]
d={}
for i in nums:
    if i%3==0:
        d[i]=[]
        if i%2==0:
            d[i]="even"
        else:
            d[i]="odd"
print(d)
```
