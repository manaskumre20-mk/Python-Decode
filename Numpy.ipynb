```python
#wapp to print one dimensional array with the help of list (take user input)
l=[]
for i in range(5):
    n=int(input("enter any number: "))
    l.append(n)
arr=np.array(l)
print(arr)
print(type(arr))
```

```python
l1=[1,6,4,3,5]
l2=[6,1,3,4,2]
l=[]
for i,j in zip(l1,l2):
    if (i**j)%2==0:
        l.append(i**j)
print(l)
```
